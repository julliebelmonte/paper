## 2. Funções de carregamento e limpeza

```python
def load_and_unify(dfs: dict, column_mapping: dict) -> pd.DataFrame:
    """
    Renomeia as colunas de cada ano conforme o mapeamento e concatena tudo.
    
    Args:
        dfs: dict {ano (str): DataFrame}
        column_mapping: dict {ano: {novo_nome: nome_original}}

    Returns:
        DataFrame unificado com coluna 'ano_pesquisa'.
    """
    frames = []
    for year, df in dfs.items():
        mapping = column_mapping[year]
        year_df = df[list(mapping.values())].rename(columns={v: k for k, v in mapping.items()})
        year_df.insert(0, 'ano_pesquisa', year)
        frames.append(year_df)
    return pd.concat(frames, ignore_index=True)


def clean_genero(df: pd.DataFrame) -> pd.DataFrame:
    """Mantém apenas respondentes que se identificam como Masculino ou Feminino."""
    return df[df['genero'].isin(['Masculino', 'Feminino'])].copy()


def clean_nivel_ensino(df: pd.DataFrame) -> pd.DataFrame:
    """Padroniza 'Pós-graduação' → 'Especialização Lato Sensu', remove 'Prefiro não informar' e define categoria ordenada."""
    s = df['nivel_ensino'].replace({
        'Pós-graduação': 'Especialização Lato Sensu',
        'Prefiro não informar': np.nan,
    })
    df['nivel_ensino'] = pd.Categorical(s, categories=NIVEL_ENSINO_ORDEM, ordered=True)
    return df


def clean_faixa_salarial(df: pd.DataFrame) -> pd.DataFrame:
    """Normaliza espaços, aplica mapeamento de faixas e define categoria ordenada."""
    s = df['faixa_salarial'].str.replace(r'\s+', ' ', regex=True).str.strip().replace(FAIXA_SALARIAL_MAP)
    df['faixa_salarial'] = pd.Categorical(s, categories=FAIXA_SALARIAL_ORDEM, ordered=True)
    return df


def clean_cargo(df: pd.DataFrame) -> pd.DataFrame:
    """Consolida cargos em categorias macro. Mantém valor original quando não há mapeamento."""
    df['cargo'] = df['cargo'].replace(CARGO_MAP)
    return df


def clean_nivel_senioridade(df: pd.DataFrame) -> pd.DataFrame:
    """
    Promove respondentes com gestor==1 para 'Gestor' e aplica
    categoria ordinal [Júnior, Pleno, Sênior, Gestor].
    """
    nivel = df['nivel'].copy()
    nivel[df['gestor'] == 1.0] = 'Gestor'
    df['nivel'] = pd.Categorical(nivel, categories=NIVEL_SENIORIDADE_ORDEM, ordered=True)
    return df


def clean_area_formacao(df: pd.DataFrame) -> pd.DataFrame:
    """Consolida áreas de formação em grupos padronizados."""
    df['area_formacao'] = df['area_formacao'].replace(AREA_FORMACAO_MAP)
    return df


def clean_situacao(df: pd.DataFrame) -> pd.DataFrame:
    """Agrupa situações de trabalho e remove linhas com 'Prefiro não informar'."""
    df['situacao'] = df['situacao'].replace(SITUACAO_MAP)
    return df[df['situacao'].notna()].copy()


def run_pipeline(dfs: dict) -> pd.DataFrame:
    """
    Pipeline completo de preprocessing.
    Encadeia todas as etapas de limpeza e retorna o DataFrame final.
    """
    df = load_and_unify(dfs, COLUMN_MAPPING)
    df = clean_genero(df)
    df = clean_nivel_ensino(df)
    df = clean_faixa_salarial(df)
    df = clean_cargo(df)
    df = clean_nivel_senioridade(df)
    df = clean_area_formacao(df)
    df = clean_situacao(df)
    return df.reset_index(drop=True)

 # Injeta bloco de IA (apenas 2024; NaN para os demais anos)
    
df = merge_ia_block(df, dfs['2024'])
    
return df.reset_index(drop=True)
```
